```python
class Queue:
    def __init__(self):
        self.data = []

    def enqueue(self, item):
        self.data.append(item)   # O(n) worst-case, usually O(1)

    def dequeue(self):
        # return self.data.pop()
        ret_val = self.data[0]
        del self.data[0] # O(n), because need to move self.data[1:] to self.data[0:]
        return ret_val    
```

```python
class CircularQueue:
    def __init__(self, size):
        self.data = [None] * size
        self.size = size
        self.start = 0
        self.end = -1

    def enqueue(self, item):
        if self.end == self.size-1:
            self.data[0] = item
            self.end = 0

        else:
            self.data[self.end+1] = item
            self.end += 1


    def dequeue(self):
        ret_val = self.data[self.start]
        self.data[self.start] = None
        if self.start == self.size-1:
            self.start = 0
        else:
            self.start += 1
            
        return ret_val    
    
    def __str__(self):
        line1 = "index\t"
        for i in range(self.size):
            line1 += str(i)
            if i != self.size-1:
                line1 += "\t"
            else:
                line1 += '\n'

        line2 = "     \t"
        for i in range(self.size):
            if self.data[i]!= None:
                line2 += str(self.data[i])
            else:
                line2 += " "

            if i != self.size-1:
                line2 += "\t"
            else:
                line2 += '\n'

        line3 = "Begin at index: " + str(self.start) + "\n"
        line4 = "End at index: " + str(self.end) + "\n"

        return line1+line2+line3+line4
    
    def __lt__(self, other):
        if self.data == other.data:
            return False
        
        for i in range(max(self.size, other.size)):
            if self.data[i] and other.data[i]:
                if self.data[i] > other.data[i]:
                    return False
            
        
        return True
        

        

    
        
```

```python
q = CircularQueue(6)
print(q)
q.enqueue(1)
print(q)
q.enqueue(2)
print(q)
q.dequeue()
print(q)
q.enqueue(3)
print(q)
q.dequeue()
print(q)
q.enqueue(1)
q.enqueue(1)
q.enqueue(1)
q.enqueue(1)
q.enqueue(1)
print(q)
q.dequeue()
print(q)

q.dequeue()
print(q)
q.dequeue()
print(q)
q.dequeue()
print(q)
q.dequeue()
print(q)
q.dequeue()
print(q)
q.enqueue(1)
print(q)
```
